```python
z=x
```

```python
def gaussian_pdf(z,c,lam,mu):
  return c*np.exp(-lam*(z-mu)**2)
```

```python
from scipy.optimize import curve_fit
```

```python
y_data,bin_edges=np.histogram(z,bins=100,density=True)
```

for center point

```python
x_data = (bin_edges[:-1] + bin_edges[1:]) / 2
```

```python
guess=[max(y_data),0.001,np.mean(z)]
popt,_=curve_fit(gaussian_pdf,x_data,y_data,guess)
```

```python
c, lam, mu = popt
```

```python
print(f"c={c}, lam={lam}, mu={mu}")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x_data, y_data, label='Data')
plt.plot(x_data, gaussian_pdf(x_data, *popt), 'r', label='Fit')
plt.legend()
plt.show()
```
